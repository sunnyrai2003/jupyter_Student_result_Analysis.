```python
# gb groupby using and find parent-educ number in  maths, reading and writing
gb=df.groupby("ParentEduc").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})gb
```

```python
gb
```

```python
# heatmap in seborn..
# this heatmap show hose parent educated
```

```python
plt.figure(figsize=(4,4))
sns.heatmap(gb, annot=True)
plt.title("Retationship between Parent Education and Student's Score")
plt.show()
```

```python
# from the above chart we have conclude that there is no/negligible importon the student's score due to their parent's matrial status
```

```python
groupA=df.loc[(df['EthnicGroup']=='group A')].count()
groupB=df.loc[(df['EthnicGroup']=='group B')].count()
groupC=df.loc[(df['EthnicGroup']=='group C')].count()
groupD=df.loc[(df['EthnicGroup']=='group D')].count()
groupE=df.loc[(df['EthnicGroup']=='group E')].count()
```
